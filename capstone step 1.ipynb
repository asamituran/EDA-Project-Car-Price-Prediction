```python
def df_nans(df, limit):
    missing = df.isnull().sum()*100 / df.shape[0]
    return missing.loc[lambda x : x >= limit]
```

**I defined 3 user defined functions to get the numeric parts of every different fuel consumption type**

```python
def cons_avg(x):
    
    if type(x)==list:
        if x[0] != []: return x[0][0]
        else: return None
  
    else:
        return None
```

```python
def cons_city(x):
    
    if type(x)==list and len(x)==3:
        if x[1] != []: return x[1][0]
        else: return None
        
    else:
        return None
```

```python
def cons_country(x):
    
    if type(x)==list and len(x)==3:
        if x[2] != []: return x[2][0]
        else: return None
        
    else:
        return None
```

```python
df["fuel_consumption"].apply(cons_avg)
```

```python
df['cons_avg'] = df["fuel_consumption"].apply(cons_avg).str.extract('(\d{1,2}.\d|\d{1,3})').astype("float")
df['cons_avg']       
```

```python
df['cons_city'] =  df["fuel_consumption"].apply(cons_city).str.extract('(\d{1,2}.\d|\d{1,3})').astype("float")         
```

```python
df['cons_country'] = df["fuel_consumption"].apply(cons_country).str.extract('(\d{1,2}.\d|\d{1,3})').astype("float")  
```
